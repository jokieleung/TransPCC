```python
from depoco.plot_results import *
path = '../experiments/results/kitti/original_depoco/'
files = sorted(glob.glob(path+'*.pkl'))
print(files)

transPCC_path = '../experiments/results/kitti/transPCC'
transPCC_files = sorted(glob.glob(transPCC_path+'*.pkl'))
print(transPCC_files)

f, ax = plt.subplots(2, 2)
ax = ax.flatten()
f.suptitle('Kitti Results')
genPlots(files, f, ax, draw_line=True, label='depoco', x_key='bpp')
genPlots(transPCC_files, f, ax, draw_line=True, label='transPCC', x_key='bpp')
for a in ax:
    a.grid()
    # a.set_ylim([0,None])
    a.legend()
plt.tight_layout()
plt.show()
```
